```python
import numpy as np
import pandas as pd

from IPython.display import display
```

```python
dataset_path = "../data/loan_data_2007_2014.csv"
df = pd.read_csv(dataset_path)
```

```python
df['loan_status'].value_counts()
```

```python
bad_statuses = [
    'Charged Off', 'Default', 'Late (31-120 days)', 
    'Late (16-30 days)', 'In Grace Period', 
    'Does not meet the credit policy. Status:Charged Off'
]

df['target'] = df['loan_status'].isin(bad_statuses).astype(int)

df.head()
```

```python
# Handle Kolom Nan
missing_percentage = df.isnull().mean() * 100

cols_to_drop = missing_percentage[missing_percentage > 50].index.tolist()

print(f"Jumlah kolom yang dibuang karena >50% kosong: {len(cols_to_drop)}")
print("Daftar kolom yang dibuang:", cols_to_drop)

df_clean = df.drop(columns=cols_to_drop)

print(f"Dimensi data setelah kolom kosong dibuang: {df_clean.shape}")
df_clean.head()
```

```python
df_clean.info()
```

```python
irrelevant_columns = ['Unnamed: 0', 'id', 'member_id', 'url', 'emp_title', 'title',
                      'issue_d', 'total_pymnt', 'recoveries', 'collection_recovery_fee', 'last_pymnt_amnt',
                      'total_pymnt_inv', 'total_rec_prncp', 'total_rec_int', 'total_rec_late_fee',
                      'last_pymnt_d', 'next_pymnt_d', 'last_credit_pull_d',
                      'out_prncp', 'out_prncp_inv',
                      'zip_code', 'addr_state',
                      'loan_status']

df_clean = df_clean.drop(columns=irrelevant_columns)

nunique_colums = [col for col in df_clean.columns if df[col].nunique() == 1]

print(f"Kolom zero variance : {nunique_colums}")
display(df_clean[nunique_colums].value_counts())

df_clean = df_clean.drop(columns=nunique_colums)

df_clean.info()
```

```python
df_clean['pymnt_plan'].value_counts()
```

```python
df_clean = df_clean = df_clean.drop(columns='pymnt_plan')
df_clean.info()
```

```python
df_clean['grade'].value_counts()
```

```python
df_clean['sub_grade'].value_counts()
```

```python
df_clean = df_clean.drop(columns='grade')
df_clean.info()
```

```python
df_clean['earliest_cr_line'].value_counts()
```

```python
split_date = df_clean['earliest_cr_line'].str.split('-', expand=True)
months_str = split_date[0]
years_short = split_date[1].astype('Int64')

years_full = (2000 + years_short).mask(years_short > 26, 1900 + years_short)

month_map = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6, 
    'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
}
month_nums = months_str.map(month_map).astype('Int64')

df_clean['mths_since_earliest_cr_line'] = (2026 - years_full) * 12 + (6 - month_nums)

df_clean = df_clean.drop(columns='earliest_cr_line')
df_clean.info()
```

```python
df_clean['mths_since_earliest_cr_line'].head()
```

## Cleaning NaN Rows

```python
df_clean.isna().sum()
```

```python
nan_ratio = df_clean.isna().mean()

cols_low_nan = nan_ratio[(nan_ratio > 0) & (nan_ratio < 0.01)].index

display(cols_low_nan)

df_clean = df_clean.dropna(subset=cols_low_nan)
df_clean.info()
```

```python
df_clean.isna().sum()
```

```python
df_clean['emp_length'].value_counts()
```

```python
df_clean['emp_length'] = df_clean['emp_length'].fillna(-1)

emp_length_map = {
    -1: -1,
    '< 1 year': 0,
    '2 years': 2,
    '3 years': 3,
    '5 years': 5,
    '1 year': 1,
    '4 years': 4,
    '7 years': 7,
    '6 years': 6,
    '8 years': 8,
    '9 years': 9,
    '10+ years': 10,
}

df_clean['emp_length'] = df_clean['emp_length'].map(emp_length_map).astype('int32')

df_clean['emp_length'].value_counts()
```

```python
df_clean['tot_coll_amt'] = df_clean['tot_coll_amt'].fillna(0)

df_clean['tot_cur_bal'] = df_clean['tot_cur_bal'].fillna(df_clean['tot_cur_bal'].median())
df_clean['total_rev_hi_lim'] = df_clean['total_rev_hi_lim'].fillna(df_clean['total_rev_hi_lim'].median())
```
